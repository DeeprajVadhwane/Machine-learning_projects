# What is the BMI and How Insurance Companies work?


  Insurance companies, including life insurance companies like LIC (Life Insurance Corporation of India), often use an individual's Body Mass Index (BMI) as one of many factors when determining insurance premiums and coverage. BMI is a numerical value calculated from an individual's height and weight and is used as an indicator of body fatness. It's calculated using the following formula:

  BMI = (Weight in kilograms) / (Height in meters)²

  The insurance company uses BMI as a health assessment tool because it can be indicative of an individual's overall health and the potential risks they may pose as an insured person. Here's how BMI data may be used in insurance:

  Pricing Premiums: Individuals with a high BMI, especially in the obese or morbidly obese category, may be considered higher risk by insurance companies. This is because obesity is associated with a higher likelihood of various health conditions, such as diabetes, heart disease, and hypertension. As a result, individuals with higher BMIs may be charged higher premiums for life insurance or health insurance.

  Underwriting Decisions: Insurance underwriters may use BMI data along with other health-related information (such as medical history, blood pressure, cholesterol levels, and family medical history) to determine if an applicant is insurable and at what rate. If an applicant has a high BMI and other health risk factors, they may be offered insurance at a higher premium or could even be denied coverage.

  Policy Limits and Exclusions: Depending on an individual's BMI and overall health, insurance policies may come with certain limitations or exclusions. For example, a policy might exclude coverage for obesity-related health conditions or surgeries if an individual has a very high BMI.

When considering BMI (Body Mass Index) for insurance purposes, both a high and low BMI can have implications, and insurance companies may assess individuals differently depending on whether their BMI is high or low. Here's what insurance companies typically consider for high and low BMIs:

1. **High BMI (Overweight or Obese):**
   - **Premiums:** Individuals with a high BMI, especially in the obese category, may face higher insurance premiums. This is because obesity is associated with an increased risk of various health conditions, which can lead to higher medical costs and more frequent insurance claims.
   - **Underwriting:** Insurance underwriters may request additional medical information, such as recent medical exams, to assess an applicant's health more thoroughly if they have a high BMI. The underwriter will consider other health factors and medical history to determine the applicant's insurability and premium rates.
   - **Exclusions:** There might be exclusions or limitations in the insurance policy related to health conditions associated with obesity. These may affect coverage for obesity-related illnesses and surgeries.

2. **Low BMI (Underweight or Very Thin):**
   - **Premiums:** In some cases, individuals with a very low BMI might also face higher premiums because being underweight can be associated with health risks and potential medical complications.
   - **Underwriting:** Insurance underwriters may evaluate the health of underweight individuals more closely, looking for signs of underlying health conditions or malnutrition. If an applicant is underweight due to an underlying health issue, it could affect their insurability or premium rates.
   - **Exclusions:** Insurance policies may have exclusions or limitations related to health conditions associated with being underweight. However, these are less common compared to exclusions related to obesity.

```python
def outlier_thershold(df,col_name,low_quantile=0.25,up_quantile=0.75):

  q1=df[col_name].quantile(low_quantile)
  q3=df[col_name].quantile(up_quantile)
  iqr=q3-q1
  low_limit=q1-1.5 * iqr
  up_limit=q3+1.5*iqr
  return low_limit,up_limit
```

```python
def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thershold(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False
```

```python
for col in df.select_dtypes(['int','float']):
    print(col, check_outlier(df, col))
```

```python
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thershold(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
```

```python
for col in df.select_dtypes(['int','float']):
    print(col, replace_with_thresholds(df, col))
```
